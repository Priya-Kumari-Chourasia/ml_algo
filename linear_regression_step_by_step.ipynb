```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=23)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
#Lets apply gradient descent assuming slope is costant m=-119.86850168
# and lets assume the starting value for intercept b = 0

y_pred = ((-119.86850168 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = -119.86850168
b = 0

loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((-119.86850168 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'black',label='OLS')
plt.plot(x,y_pred1,color='blue',label = 'b= {}'.format(b))
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size =loss_slope*lr
step_size
```

```python
b = b- step_size
b
```

```python
y_pred2 = ((-119.86850168 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='black',label='ols')
plt.plot(x,y_pred2,color='yellow',label='b={}'.format(b))
plt.plot(x,y_pred1,color='red',label = 'b it1'.format(b))
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = lr*loss_slope
step_size
```

```python
b = b- step_size
b
```

```python
y_pred3 = ((-119.86850168 * x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='black',label='ols')
plt.plot(x,y_pred3,color='green',label='b={}'.format(b))
plt.plot(x,y_pred2,color='yellow',label='b it2'.format(b))
plt.plot(x,y_pred1,color='red',label = 'b it1'.format(b))
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```
