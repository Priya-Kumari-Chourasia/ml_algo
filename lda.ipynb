```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
import plotly.express as px
```

```python
iris = load_iris()
df = pd.DataFrame(data = iris.data,columns = iris.feature_names)
df['target'] = iris.target

df.head()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
import plotly.express as px
fig = px.scatter_3d(df,x=df['feature1'],y=df['feature2'],z=df['feature3'],color=df['target'].astype('str'))
fig.update_traces(marker=dict(size=12,line=dict(width=5,color='DarkSlateGrey')),selector=dict(mode='markers'))

fig.show()
```

```python
fig = px.scatter_3d(df,x=df.iloc[:,0],y=df.iloc[:,1],z=df.iloc[:,2],color=df['target'].astype('str'))
fig.update_traces(
    marker=dict(
        size=8,
        line=dict(width=4, color='DarkSlateGrey')
    ),
    selector=dict(mode='markers')
)
fig.show()
```

```python
df.sample(5)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
x_train_lda = lda.fit_transform(x_train_scaled,y_train)
x_test_lda = lda.transform(x_test_scaled)
```

```python
lda
```

```python
fig = px.scatter(x=x_train_lda[:,0],y=x_train_lda[:,1],color=y_train.astype(str),labels={'x':'LD1','y':'LD2'})
fig.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

classifier = LogisticRegression(random_state=42)
classifier.fit(x_train_lda,y_train)
y_pred = classifier.predict(x_test_lda)
```

```python
# 1. Print the Classification Report (Precision, Recall, F1-Score)
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("-" * 50)

# 2. Print the Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# 1. Create a DataFrame to compare individual predictions with real targets
comparison_df = pd.DataFrame({
    'Actual Target': y_test,
    'Predicted Target': y_pred
})

# 2. Add a column to instantly see if the model got it right
comparison_df['Correct?'] = comparison_df['Actual Target'] == comparison_df['Predicted Target']

# 3. (Optional) Map the numeric categories (0, 1, 2) back to the actual Iris flower names
target_names = iris.target_names
comparison_df['Actual Name'] = [target_names[i] for i in y_test]
comparison_df['Predicted Name'] = [target_names[i] for i in y_pred]

# Display the side-by-side comparison
comparison_df
```
