```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
x,y = make_classification(n_samples=200,n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(x,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=3)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=3)
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
print(rf.estimators_)
print(f"Number of estimators: {len(rf.estimators_)}")
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[10])
plt.show()
```
